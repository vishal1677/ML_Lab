```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.DataFrame([[0.1,0.6],[0.15,0.71],[0.08,0.9],[0.16,0.85],[0.2,0.3],[0.25,0.5],[0.24,0.1],[0.3,0.2]],columns=['X','Y'])
df
```

```python
x=df[['X','Y']]
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['X'],y=df['Y'])
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=2,init=[[0.10,0.60],[0.3,0.2]],n_init=100)
labels=km.fit_predict(x)
labels
```

```python
cen=km.cluster_centers_
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['X'],y=df['Y'],c=labels)
plt.scatter(cen[:,0],cen[:,1],s=100,c='r')
plt.show()
```
