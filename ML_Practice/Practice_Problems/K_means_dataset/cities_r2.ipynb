```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('cities_r2.csv')
df.head()
```

```python
df.shape
```

```python
x=df[['total_graduates']]
```

```python
from sklearn.cluster import KMeans
```

```python
wsse=[]
for i in range(1,16):
    km=KMeans(n_clusters=i)
    km.fit_predict(x)
    wsse.append(km.inertia_)
wsse
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,16),wsse,marker='.',color='r')
plt.grid()
plt.show()
```

```python
silh=[]
from sklearn.metrics import silhouette_score
for i in range(2,16):
    km=KMeans(n_clusters=i,n_init=100)
    labels=km.fit_predict(x)
    silh.append(silhouette_score(x,labels))
silh
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(2,16),silh,marker='.',color='r')
plt.grid()
plt.show()
```

```python
km=KMeans(n_clusters=2)
labels=km.fit_predict(x)
labels
```

```python
centr=km.cluster_centers_
```
