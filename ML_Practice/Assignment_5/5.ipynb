```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['Spending Score (1-100)'],y=df['Annual Income (k$)'])
plt.title('Unclustered data')
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
x=df.iloc[:,3:5]
print(x)
```

```python
wsse=[]
for i in range(1,16):
    km=KMeans(n_clusters=i)
    km.fit_predict(x)
    wsse.append(km.inertia_)
wsse
```

```python
#Elbow method
plt.figure(figsize=(10,6))
plt.plot(range(1,16),wsse,color='r',marker='.')
plt.grid()
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
silh=[]
for i in range(2,16):
    km=KMeans(i)
    labels=km.fit_predict(x)
    silh.append(silhouette_score(x,labels))
silh
```

```python
#Elbow method
plt.figure(figsize=(10,6))
plt.plot(range(2,16),silh,color='r',marker='.')
plt.grid()
plt.show()
```

```python
km=KMeans(n_clusters=5)
labels=km.fit_predict(x)
labels
```

```python
centr=km.cluster_centers_
centr
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x=df['Spending Score (1-100)'],y=df['Annual Income (k$)'],c=labels)
plt.scatter(centr[:,0],centr[:,1],c='r',s=100)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
sns.scatterplot(x=df['Spending Score (1-100)'],y=df['Annual Income (k$)'])
plt.title('Unclustered data')
plt.subplot(1,2,2)
sns.scatterplot(x=df['Spending Score (1-100)'],y=df['Annual Income (k$)'],c=labels)
plt.scatter(centr[:,0],centr[:,1],c='r',s=100)
plt.title('Clustered data')
plt.show()
```

```python
#hierarchiel clustering 
import  scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(10,6))
sch.dendrogram(sch.linkage(x,method='ward'))
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
ag=AgglomerativeClustering(n_clusters=5)
labels2=ag.fit_predict(x)
labels2
```

```python
plt.figure(figsize=(10,6))
plt.subplot(1,2,1)
sns.scatterplot(x=df['Spending Score (1-100)'],y=df['Annual Income (k$)'],c=labels2)
plt.title('Hierarchiel Clustered data')
plt.subplot(1,2,2)
sns.scatterplot(x=df['Spending Score (1-100)'],y=df['Annual Income (k$)'],c=labels)
plt.scatter(centr[:,0],centr[:,1],c='r',s=100)
plt.title('Kmeans Clustered data')
plt.show()
```
