```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import association_rules,apriori
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Market_Basket_Optimisation.csv')
df.head()
```

```python
data = []
for index,row in df.iterrows():
    temp=[]
    for j in row:
        if pd.notna(j):
            temp.append(j)
    data.append(temp)
```

```python
data[:10]
```

```python
te=TransactionEncoder()
x=te.fit_transform(data)
x
```

```python
df1=pd.DataFrame(x,columns=te.columns_)
df1.head()
```

```python
df1.shape
```

```python
freq_pat=apriori(df1,min_support=0.01,use_colnames=True)
freq_pat
```

```python
rules=association_rules(freq_pat,metric='confidence',min_threshold=0.01)
```

```python
rules
```

```python
rules=rules[['antecedents','consequents','support','confidence','lift']]
rules
```
