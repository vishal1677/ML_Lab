```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('mushrooms.csv')
df.head()
```

```python
df.shape
```

```python
df=pd.get_dummies(df,prefix_sep='_')
df.head()
```

```python
df.shape
```

```python
freq_pat=apriori(df,min_support=0.4,use_colnames=True)
freq_pat
```

```python
rules=association_rules(freq_pat,metric='confidence',min_threshold=0.4)
rules
```

```python
rules=rules[['antecedents','consequents','support','confidence','lift']]
rules
```

```python
rules[rules['antecedents'] == {'odor_n'}]['consequents']
```
