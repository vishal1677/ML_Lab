```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
import csv
```

```python
dataset=[]
with open('Market_Basket_Optimisation.csv') as file:
    reader=csv.reader(file,delimiter=',')
    for read in reader:
        dataset+=[read]
dataset[0:10]
```

```python
te=TransactionEncoder()
x=te.fit_transform(dataset)
x
```

```python
import pandas as pd
```

```python
len(te.columns_)
```

```python
df=pd.DataFrame(x,columns=te.columns_)
df.head()
```

```python
freq_item=apriori(df,min_support=0.01,use_colnames=True)
freq_item
```

```python
rules=association_rules(freq_item,metric='confidence',min_threshold=0.1)
rules
```

```python
rules=rules[['antecedents','consequents','support']]
rules
```
