```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df=pd.read_csv('Mall_Customers.csv')
df.head()
```

```python
x=df.iloc[:,3:5]
print(x)
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=x['Annual Income (k$)'],y=x['Spending Score (1-100)'])
plt.title('Unclustered data')
plt.show()
```

```python
#find the best no of clusters using silhoute and elbow method
from sklearn.cluster import KMeans,AgglomerativeClustering
km=KMeans(n_clusters=3)
labels=km.fit_predict(x)
labels
```

```python
sse=[]
for i in range(1,16):
    km=KMeans(n_clusters=i)
    km.fit_predict(x)
    sse.append(km.inertia_)
sse    
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(1,16),sse,marker='.',color='red')
plt.grid()
plt.xlabel('No of clusters')
plt.ylabel('SSE')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
silh=[]
for i in range(2,16):
    km=KMeans(n_clusters=i)
    labels=km.fit_predict(x)
    silh.append(silhouette_score(x,labels))
silh    
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(2,16),silh,marker='.',color='red')
plt.grid()
plt.xlabel('No of clusters')
plt.ylabel('Silhoute score')
plt.show()
```

```python
km=KMeans(n_clusters=5)
labels=km.fit_predict(x)
labels
centr=km.cluster_centers_
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=x['Annual Income (k$)'],y=x['Spending Score (1-100)'],c=labels)
plt.scatter(centr[:,0],centr[:,1],s=100,color='r')
plt.show()
```

```python
#Aglomarative claustering 
ag=AgglomerativeClustering(n_clusters=5)
ag_labels=ag.fit_predict(x)
ag_labels
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.scatterplot(x=x['Annual Income (k$)'],y=x['Spending Score (1-100)'])
plt.title('Unclustered')
plt.subplot(1,2,2)
sns.scatterplot(x=x['Annual Income (k$)'],y=x['Spending Score (1-100)'],c=labels)
plt.scatter(centr[:,0],centr[:,1],color='r')
plt.title('Clustered')
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.scatterplot(x=x['Annual Income (k$)'],y=x['Spending Score (1-100)'],c=labels)
plt.scatter(centr[:,0],centr[:,1],color='r')
plt.title('K means')
plt.subplot(1,2,2)
sns.scatterplot(x=x['Annual Income (k$)'],y=x['Spending Score (1-100)'],c=ag_labels)
plt.title('Aglomarative')
```
